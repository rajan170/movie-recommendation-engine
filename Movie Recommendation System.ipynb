```python
import ast
'''Converts dictionary to lists of lists'''
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
import ast
'''Converts dictionary to lists of lists'''
def convert2(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<4:  #to select first 4 characters and names only
#             L.append(i['name']+' : '+i['character'])
              L.append(i['name']+' '+i['character'])
              counter+=1
        else:
            break
    return L    
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert2)
movies_df['cast']
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']  == 'Director':
            L.append(i['name'])
            break
    return L
```

## Stemming(Truncating words to their origin words)

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem("action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthingtonjakesully zoesaldananeytiri sigourneyweaverdr.graceaugustine stephenlangcol.quaritch jamescameron in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization.")
```

```python
def stem(text):
    L=[]
    for i in text.split():
        L.append(ps.stem(i)) 
    return " ".join(L)     
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

## Vectorizing tags

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

### We need to find distance between similar words but we can't use euclidean distance or any other distance measure because we are dealing with multi-dimensional space and normal distance are'nt reliable in multi-dim space. So, we will use cosine angle between two vectors to find distance.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

### Distance of every vector to each other and themself. Which means how much one movie is related to other, in this case, the amount every movie is related to each other.

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
#Top 5 movies that are most closely related to the first movie
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
#         print(new_df.iloc[i[0]].title, i[0] )
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
